I have chosen an approximately 300km wide coastal region with the far left encompassing Cape Agulhas. I chose this region because I will be extending an existing ocean model (from Durban to Port Elizabeth) from Port Elizabeth to Cape Agulhas in my dissertation so I want to learn as much as I can about the oceanography of this region. 

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import pandas as pd
import numpy as np
import cartopy.feature as cfeature
from matplotlib import ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
from matplotlib.ticker import MultipleLocator
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'projection': ccrs.PlateCarree()})
snippet_extent = [19.344727, 22.020996, -35.414124, -34.303514] 
south_africa_extent = [16, 33, -35.5, -22]
cape_agulhas = {'lat': -34.833, 'lon': 20.000}
witsand = {'lat': -34.39422638518583, 'lon': 20.842622033254703}

#first map: chosen region map with cape agulhas labelled
axes[0].set_extent(snippet_extent, crs=ccrs.PlateCarree())
axes[0].coastlines()
axes[0].add_feature(cfeature.LAND)
axes[0].add_feature(cfeature.OCEAN)
gl_0 = axes[0].gridlines(draw_labels=True, linestyle='--', alpha=0.5)

# gridlines formatting
gl_0.xlocator = MultipleLocator(0.5)  # Longitude lines every 10°
gl_0.ylocator = MultipleLocator(0.5)   # Latitude lines every 5°
gl_0.xformatter = LONGITUDE_FORMATTER
gl_0.yformatter = LATITUDE_FORMATTER
gl_0.top_labels = False
gl_0.right_labels = False

axes[0].set_title("Coastal region surrounding Cape Agulhas", fontweight='bold')
#plot cape agulhas point
axes[0].plot(cape_agulhas['lon'], cape_agulhas['lat'], marker='o', color='red', markersize=3, transform=ccrs.PlateCarree())
axes[0].text(cape_agulhas['lon'] + 0.1, cape_agulhas['lat'], 'Cape Agulhas', transform=ccrs.PlateCarree())
#add witsand
axes[0].plot(witsand['lon'], witsand['lat'], marker='o', color='red', markersize=3, transform=ccrs.PlateCarree())
axes[0].text(witsand['lon'] + 0.05, witsand['lat']-0.1, 'Witsand', transform=ccrs.PlateCarree())


#second map: map of south africa with cape agulhas labelled
axes[1].set_extent(south_africa_extent, crs=ccrs.PlateCarree())
axes[1].coastlines()
axes[1].add_feature(cfeature.LAND)
axes[1].add_feature(cfeature.OCEAN)
gl_1 = axes[1].gridlines(draw_labels=True, linestyle='--', alpha=0.5)

# gridlines formatting
gl_1.top_labels = False
gl_1.left_labels = False

axes[1].set_title("Cape Agulhas coastal region relative to South Africa", fontweight='bold')
#plot cape agulhas point
axes[1].plot(cape_agulhas['lon'], cape_agulhas['lat'], marker='o', color='red', markersize=3, transform=ccrs.PlateCarree())
axes[1].text(cape_agulhas['lon'] + 0.2, cape_agulhas['lat']-0.2, 'Cape Agulhas', transform=ccrs.PlateCarree())



fig.text(0.5, 0.02,"Projection: Cylindrical Equidistant", ha='center', fontsize=9, color='grey')
plt.show()
fig.savefig('map_of_cape_agulhas_coast.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
ocean_color_dataset = xr.open_dataset("../ESACCI-OC-MAPPED-CLIMATOLOGY-1M_MONTHLY_4km_PML_CHL-fv5.0.nc", engine="netcdf4")

print(ocean_color_dataset)
print(ocean_color_dataset['chlor_a'].attrs)

#normal conc version
chlorophyll_conc_xarray = ocean_color_dataset['chlor_a'].sel(
    lon=slice(19.344727, 22.020996),
    lat=slice(-34.303514, -35.414124)
)
sorted_chlorophyll_conc_xarray = chlorophyll_conc_xarray.sortby('time')

#log transformed conc version
ocean_color_dataset['log10_chlor_a'] = np.log10(ocean_color_dataset['chlor_a'])
chlorophyll_log10_conc_xarray = ocean_color_dataset['log10_chlor_a'].sel(
    lon=slice(19.344727, 22.020996),
    lat=slice(-34.303514, -35.414124)
)
sorted_chlorophyll_log10_conc_xarray = chlorophyll_log10_conc_xarray.sortby('time')

```

```python
#LOG TRANSFORMED VERSION - variations are clearer with log transform

#extent of snippet from GMRRT:
#West: 19.344727
#East: 22.020996
#South: -35.414124
#North: -34.303514

# mean over time
chlorophyll_log10_mean_over_time = chlorophyll_log10_conc_xarray.mean(dim='time')


fig = plt.figure(figsize=(12,8))
axis = plt.axes(projection=ccrs.PlateCarree())
axis.set_extent([19.344727, 22.020996, -35.414124, -34.303514], crs=ccrs.PlateCarree())
axis.coastlines()

chlorophyll_log10_mean_over_time.plot.pcolormesh(
    ax=axis,
    transform=ccrs.PlateCarree(),
    cmap='viridis',
    add_colorbar=True,
    cbar_kwargs={'label': 'Chlorophyll-a (log₁₀[mg/m³])'}
)

gl = axis.gridlines(draw_labels=True, linestyle='--', alpha=0.5)
gl.top_labels = False
gl.right_labels = False
axis.set_title("Mean Annual Chlorophyll-a Concentration (log₁₀[mg/m³]) for Cape Agulhas Region\nbetween September 1997-August 1998", fontweight='bold')
fig.text(0.45, 0.02, 'Projection: Cylindrical Equidistant\nData: ESA Ocean Colour CCI', ha='center', fontsize=10, color='gray')

plt.show()
fig.savefig('mean_annual_chloro_conc.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
#LOG TRANSFORMED VERSION-variations much clearer with this version

# Create figure and 12 subplots
fig, axes = plt.subplots(3, 4, figsize=(18, 12),
                         subplot_kw={'projection': ccrs.PlateCarree()})
axes = axes.flatten()

# for consistent color scaling
vmin = sorted_chlorophyll_log10_conc_xarray.min().item()
vmax = sorted_chlorophyll_log10_conc_xarray.max().item()

# loop to plot each months data on a subplot
for i in range(12):
    ax = axes[i]
    ax.set_extent([19.344727, 22.020996, -35.414124, -34.303514], crs=ccrs.PlateCarree())
    ax.coastlines()
    
    # same color scale
    im = sorted_chlorophyll_log10_conc_xarray.isel(time=i).plot.pcolormesh(
        ax=ax,
        transform=ccrs.PlateCarree(),
        cmap='viridis',
        vmin=vmin,
        vmax=vmax,
        add_colorbar=False
    )

    formatted_date = pd.to_datetime(sorted_chlorophyll_log10_conc_xarray['time'].values[i]).strftime('%B %Y')
    ax.set_title(f"{formatted_date}")

#making space for color bar
plt.subplots_adjust(right=0.9)    

# vertical colorbar on the right
cbar_ax = fig.add_axes([0.92, 0.15, 0.015, 0.7])
cbar = fig.colorbar(im, cax=cbar_ax)
cbar.set_label('Chlorophyll-a (log₁₀[mg/m³])')


fig.suptitle("Monthly Chlorophyll-a Concentrations (log₁₀[mg/m³]) around the Cape Agulhas region\nbetween September 1997 - August 1998", fontweight='bold', fontsize=16)
fig.text(0.5, 0.1, 'Projection: Cylindrical Equidistant\nData: ESA Ocean Colour CCI', ha='center', fontsize=11, color='gray')

plt.show()
fig.savefig('monthly_chloro_conc.png', dpi=300, bbox_inches='tight', transparent=True)
```
